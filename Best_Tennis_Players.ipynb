```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
world = pd.read_csv('Tennis_World_Number_1s.csv')
```

```python
world.drop('Unnamed: 7', axis=1, inplace=True)
```

```python
world[['Careers_Wins', 'Careers_Losses']] = world.Careers_Wins_Losses.str.split('-', expand=True).astype(float)
```

```python
world[['Tennis_Player', 'Careers_Wins_Losses', 'Careers_Wins', 'Careers_Losses']].head()
```

```python
world['Total_Professional_matches'] = world.Careers_Wins + world.Careers_Losses
world['Win_Percentage'] = (world.Careers_Wins / world.Total_Professional_matches) * 100
```

```python
world[['Tennis_Player', 'Total_Professional_matches', 'Win_Percentage']].sort_values('Win_Percentage', ascending=False)
```

```python
world[['Tennis_Player', 'Years_End_no_1', 'Win_Percentage']].sort_values(['Years_End_no_1','Win_Percentage'], ascending=False).head(10)
```

```python
Tennis_Win_Perc = world.sort_values('Win_Percentage', ascending=False)
Tennis_Win_Perc['Percent_diff'] = Tennis_Win_Perc.Win_Percentage.pct_change() * 100
Tennis_Win_Perc['Percent_diff'] = Tennis_Win_Perc.Percent_diff.fillna(0)
Tennis_Win_Perc[['Tennis_Player', 'Win_Percentage', 'Percent_diff']].head(10)
```

```python
crit1 = world.Maximum_Consectutive_Weeks_at_no_1 > 30
crit2 = world.Titles > 80
crit3 = world.Prize_Money < 100000000
```

```python
world[crit1 & crit2 & crit3].loc[:,['Tennis_Player']]
```

```python
world.plot.barh(x='Tennis_Player', y='Total_Weeks_at_No_1', figsize=(15, 10))
```

```python
world[world.Total_Weeks_at_No_1 > 50].plot.barh(x='Tennis_Player', y='Total_Weeks_at_No_1', figsize=(15, 10))
```

```python
world['Earning'] = Tennis_Win_Perc.Prize_Money / 10**6
world['Earning'].head()
```

```python
tennis_player = list(world.Tennis_Player)
x_coord = list(world.Earning)
y_coord = list(world.Win_Percentage)

plt.figure(figsize=(18, 12))

for i, player in enumerate(tennis_player):
    x = x_coord[i]
    y = y_coord[i]
    plt.scatter(x, y, marker='o', color='red', alpha=0.5, s=150)
    plt.text(x+0.5, y+0.5, player, fontsize=12)
    
plt.title('The Relationship between Win Percentage and Prize Money\nfor Former World Number No.1s', fontsize=20)
plt.xlabel('Prize_Money', fontsize=16)
plt.ylabel('Win_Percentage', fontsize=16)
plt.show()
```
